```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV file ko read karo aur column names manually set karo
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv('iris.data.csv', header=None, names=column_names)

# Dataset ka structure check karo
print("Shape:", iris.shape)  # Rows aur columns ki sankhya
print("Columns:", iris.columns)  # Column ke naam
print(iris.head())  # Pehle 5 rows dikhayein

# Scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(6, 4))
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], label=species)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.show()

# Histogram: Petal Length ka distribution
plt.figure(figsize=(6, 4))
plt.hist(iris['petal_length'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Length')
plt.show()

# Box plot: Sepal Length by Species
plt.figure(figsize=(6, 4))
species_list = iris['species'].unique()
data_to_plot = [iris[iris['species'] == species]['sepal_length'] for species in species_list]
plt.boxplot(data_to_plot, labels=species_list)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Sepal Length by Species')
plt.show()
```
